```python
df[['emp_title', 'emp_length']] = df[['emp_title', 'emp_length']].replace(np.nan, 'Not Known')
```

```python
df['desc'].replace(np.NaN,'Rental/Rented')
df['desc']
```

```python
df[['mths_since_last_delinq', 'mths_since_last_record']]=df[['mths_since_last_delinq', 'mths_since_last_record']].replace(np.nan,'Record Not Found')
df[['mths_since_last_delinq', 'mths_since_last_record']]
```

```python
df['revol_util']=df['revol_util'].replace(np.nan,df['revol_util'].mean())
df['revol_util']
```

```python
for index, row in df.iterrows():
    if pd.isnull(row['last_pymnt_d']) and pd.isnull(row['next_pymnt_d']):
        df.at[index, 'last_pymnt_d'] = 'Defaulter'
        df.at[index, 'next_pymnt_d'] = 'Defaulter'
    elif pd.isnull(row['last_pymnt_d']):
        df.at[index, 'last_pymnt_d'] = 'Unpaid'
    elif pd.isnull(row['next_pymnt_d']):
        df.at[index, 'next_pymnt_d'] = 'Paid'
```

```python
df.isnull().sum()
```

```python
df.to_csv('Modified_Data.csv')
```
